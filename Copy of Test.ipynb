#### Calculte the sum of the digits of a 3 - digit random number

```python
import random
n = random.randrange(100,1000)
print("The random number is",n)
sum=0
while n>0:
  t=n%10
  sum=sum+t
  n=n//10
print("Sum is",sum)
```

#### Determine the percntage of lowercase and upper case in a string

```python
s=input("Enter the String ")
lowsum=0
upsum=0
for i in s:
  if i.isupper():
    upsum=upsum+1
  else:
    lowsum=lowsum+1
upsum=(upsum/len(s))*100
lowsum=(lowsum/len(s))*100
print("percntage of lowercase is ",lowsum)
print("percntage of lowercase is ",upsum)
```

#### Take random 1000 numbers and seperate a list of positive and negative numbers from it

```python
NumList = []
Positive = []
Negative = []
Number = 1000
for i in range(1, Number + 1):
    value = random.randrange(-1000,1000)
    NumList.append(value)

for j in range(Number):
    if(NumList[j] >= 0):
        Positive.append(NumList[j])
    else:
        Negative.append(NumList[j])

print("Element in Positive List is : ", Positive)
print("Element in Negative List is : ", Negative)
```

#### Is is possibe to form a trinagle wiht user given points ?

```python
def check(x1, y1, x2, y2, x3, y3):
    a = (x1 * (y2 - y3) +
         x2 * (y3 - y1) +
         x3 * (y1 - y2))
    if a == 0:
        print('Not possible')
    else:
        print('Possible')
            
a,b=map(int,input("Enter X1 , Y1 ").split())   
c,d=map(int,input("Enter X2 , Y2 ").split())
e,f=map(int,input("Enter X3 , Y3 ").split())
check(a,b,c,d,e,f)
```
