### Find indices of non zero element from [1, 2, 0, 0, 4, 0]

```python
q=0
arr=[1,2,0,0,4,0]
for i in arr:
  if(i!=0):
    print(q)
    q+=1
```

### Multiply a 5 X 3 matrix by 3 X 2 matrix ( real matrix product ) 

```python
arr1=np.array([1,2,3])
arr1=np.resize(5,2)
arr2=np.array([1,2,3])
arr2=np.resize(3,2)
np.dot(arr1,arr2)
```
