**create an array consistiing of 100 rows and 10 cols filled with string values in combinations of 3**<br>
Ex = 'abc', 'ade', 'klm'

```python
import random 
cs = []
for i in np.arange(200):
    val = ''.join(random.sample(string,3))
    if val in cs:
        continue
    else:
        cs.append(val)
```

```python
cs.sort()
np.array(cs).reshape(2,97)
```

```python
np.where(cs == 'cat')
```
