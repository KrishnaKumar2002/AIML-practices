```python
#from sympy import *
expr=Eq(((x/(x+1))+((x+1)/x)),34/15)
print(solve(expr))
```

## Trignometry 

### v) Evaluate $\sqrt{\frac{1 + \sin A}{1 - \sin A}}$

`output` - $\sec A + \tan A$

```python
A=sym.symbol("A")
expr=trigsimp(sqrt((1+sin(A)/(1-sin(A)))))
print(solve(expr))
```

### vi) Given $\sqrt{\frac{1 + \tan A^2}{1 + \cot A^2}} = \left(\frac{1 - \tan A}{1 - \cot A}\right)^2$

`output` - $ \tan A^2$

```python
from sympy import *
A=sym.symbol(A)
expr=sqrt((1+tan(A)**2)/(1+cot(A)**2))
eq=Eq(expr,((1-tan(A))/(1-cot(A)))**2)
trigsimp(eq)
```
