### 4. Create an Array and find the sqrt, and sin vaules of it

```python
arr=np.array([1,2,3,4,0,-1])
print(np.sqrt(arr))
print(np.sin(arr))
print(np.cos(arr))
print(np.log(arr))
```

### 5. Create a 6d Array using ndmin and convert it to Bool Type

```python
arr=[1,2,3,4,0]
arr6d=np.array(arr,ndmin=6)
print(arr6d)
arr6d.astype('bool')
```
